# Swin-Tiny: Pretrained vs From Scratch Across Data Fractions

This notebook compares `swin_tiny_patch4_window7_224` under two initialization modes:

- `pretrained`
- `scratch`

The comparison is run on multiple training fractions:

- `100%`
- `50%`
- `25%`
- `10%`
- `5%`

Each configuration is run with multiple random seeds, and the notebook reports `mean ± std`.

Saved artifacts:

- `checkpoints/`
- `results/`
- `figures/`


```python
from pathlib import Path
import csv
import json

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

project_root = Path('.')
script_path = project_root / 'run_swin_pretrained_vs_scratch.py'
print(f'Using script: {script_path.resolve()}')
```

```python
%run ./run_swin_pretrained_vs_scratch.py --seeds 42 123 3185
```

```python
summary_path = project_root / 'results' / 'swin_tiny_pretrained_vs_scratch_summary.csv'
with summary_path.open('r', encoding='utf-8') as handle:
    rows = list(csv.DictReader(handle))

display(Markdown('## Aggregated Summary Table (`mean ± std`)'))
rows
```

```python
aggregate_json_path = project_root / 'results' / 'swin_tiny_pretrained_vs_scratch_aggregate.json'
aggregate_payload = json.loads(aggregate_json_path.read_text(encoding='utf-8'))

for row in aggregate_payload:
    display(Markdown(f"### {row['init_mode']} | {row['train_fraction']:.0%}"))
    print('Seeds:', row['seeds'])
    print('Accuracy:', f"{row['test_accuracy_mean']:.4f} ± {row['test_accuracy_std']:.4f}")
    print('Macro F1:', f"{row['test_macro_f1_mean']:.4f} ± {row['test_macro_f1_std']:.4f}")
    print('AUC:', f"{row['test_auc_mean']:.4f} ± {row['test_auc_std']:.4f}")
    print()
```

```python
json_dir = project_root / 'results'
for json_path in sorted(json_dir.glob('swin_tiny_*pct_seed*_metrics.json')):
    payload = json.loads(json_path.read_text(encoding='utf-8'))
    display(Markdown(f"### Single Run: {payload['init_mode']} | {payload['train_fraction']:.0%} | seed={payload['config']['seed']}"))
    print('Best epoch:', payload['best_epoch'])
    print('Checkpoint:', payload['checkpoint_path'])
    print('Test metrics:', payload['test_metrics'])
    print()
```

```python
figure_paths = [
    project_root / 'figures' / 'bloodmnist_swin_pretrained_vs_scratch_accuracy_auc.png',
    project_root / 'figures' / 'bloodmnist_swin_pretrained_vs_scratch_macro_f1.png',
]

for figure_path in figure_paths:
    display(Markdown(f"## {figure_path.name}"))
    image = mpimg.imread(figure_path)
    plt.figure(figsize=(12, 5))
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```
